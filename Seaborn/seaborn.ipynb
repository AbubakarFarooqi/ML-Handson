```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing sample dataset from seaborn
data = sns.load_dataset("tips")
```

```python
data
```

Scatter plot

```python
sns.scatterplot(x="total_bill",y="tip",data=data,hue='time',style='time')
plt.title("A Scatter plot in Seaborn")
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.show()
```

Line Plot

```python
sns.lineplot(x="size",y="total_bill",data=data)
plt.title("A Line plot in Seaborn")
plt.xlabel("Size")
plt.ylabel("total_bill")
plt.show()
```

CATEGORICAL PLOTS
Bar plot

Bar PLot

```python
sns.barplot(x="sex",y="total_bill",data=data)
plt.title("A Bar plot in Seaborn")
plt.xlabel("Sex")
plt.ylabel("total_bill")
plt.show()
```

BOX PLOT

```python
sns.boxplot(x="sex",y="size",data=data)
plt.title("A box plot in Seaborn")
plt.xlabel("sex")
plt.ylabel("size")
plt.show()
```

HISTOGRAM

```python
sns.histplot(data["total_bill"],bins=10)
plt.show()
```

```python
sns.histplot(data["total_bill"],bins=10,kde=True)
plt.show()
```

KDE PLOT

```python
sns.kdeplot(data["total_bill"],fill=True)
plt.show()
```

PAIR PLOT

```python
sns.pairplot(data)
```

HEAT  MAP

```python
corr = data.select_dtypes(["number"]).corr()
sns.heatmap(data=corr,annot=True,cmap="YlOrBr")
```
