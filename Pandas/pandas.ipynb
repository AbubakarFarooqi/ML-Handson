```python
#Series Data Structure: Just like a 1-D array and can have multiple types of data
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(series.shape)
```

```python
# Series with different index
data = [1,2,3,'a']
index = ['a','b','c','d']

series = pd.Series(data,index=index)
print(series)
```

```python
## DATA FRAME: A 2-d tabular data structure
data =  {
    'col1':[1,2,3],
    'col2':[2,3,None],

    'col3':["AZAN","USMAN","UMAR"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# creating dataframe by list of dictionaries
data = [
    {'col1':1,'col2':2,'col3':"AZAN"},
    {'col1':2,'col2':3,'col3':"USMAN"},
    {'col1':3,'col2':None,'col3':"UMAR"}
]

df = pd.DataFrame(data)
print(df)
```

```python
#get top n and last n rows
print(df.head(2))

print(df.tail(1))
```

```python
#Accessing data from dataframe
# if i pick a column it will give me the series

print((df['col1']))
print((type(df['col1'])))
```

```python
# data manipulations with dataframes

# adding a new column
df['new_col'] = [1,2,3]
df
```

```python
# droping column
df.drop('new_col',axis = 1,inplace=True)
df
```

```python
# Arithmetics in column
df['col1'] = df['col1']+1
df
```
