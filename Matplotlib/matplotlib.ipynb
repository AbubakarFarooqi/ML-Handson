```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

```python
##Line Plot

plt.plot(x,y)
```

```python
##Line Plot with labels and title

plt.plot(x,y)

plt.xlabel("values")
plt.ylabel("Squares")
plt.title("Value vs Square")
plt.show()
```

```python
##Line Plot with customizations

x = list(range(-25, 26))
y = [i**2 for i in x]

plt.xlabel("values")
plt.ylabel("Squares")
plt.title("Value vs Square")

plt.plot(x,y,color='Red',linestyle="--",marker="o",linewidth = 1,markersize = 5,label="y=x^2")
plt.grid(visible=True)

plt.show()
```

Multiple plots

```python
## multiple plots

plt.figure(figsize=(9,15)) # (width,height) in inches

plt.subplot(2,2,1) # (row_position, col_position, cell number)
plt.plot([1,2,3,4,5],[2,4,6,8,10])
plt.title("Plot 1")

plt.subplot(2,2,2) 
plt.plot([1,2,3,4,5],[2,4,6,8,10][::-1])
plt.title("Plot 2")

plt.subplot(2,2,3) 
plt.plot(x,y)
plt.title("Plot 3")

plt.subplot(2,2,4) 
plt.plot([1,2,3,4,5],[1,1,1,1,1])
plt.title("Plot 4")


```

BAR PLOT

```python
##BAR PLOT
x = ['a','b','c','d']
y = [4,2,5,3]

plt.bar(x,y,color='green')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("A bar plot")
plt.show()
```

HISTOGRAM
A histogram distributes data in bins and display, usually use to check distribution of data

```python
##HISTOGRAM
data = [1,1,1,1,2,3,4,5,5,5,5,6]
plt.hist(data,bins=4,color="orange",edgecolor='black')
plt.xlabel("Values Range")
plt.ylabel("count")
plt.title("A Histogram plot")
plt.show()
```

SCATTER PLOT

```python
x = [1,2,3,4]
y = [-4,2,0,5]

plt.scatter(x,y,color="green",marker="o")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("A Scatter plot")
plt.show()
```
