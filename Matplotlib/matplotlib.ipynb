```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

```python
##Line Plot

plt.plot(x,y)
```

```python
##Line Plot with labels and title

plt.plot(x,y)

plt.xlabel("values")
plt.ylabel("Squares")
plt.title("Value vs Square")
plt.show()
```

```python
##Line Plot with customizations

x = list(range(-25, 26))
y = [i**2 for i in x]

plt.xlabel("values")
plt.ylabel("Squares")
plt.title("Value vs Square")

plt.plot(x,y,color='Red',linestyle="--",marker="o",linewidth = 1,markersize = 5,label="y=x^2")
plt.grid(visible=True)

plt.show()
```
